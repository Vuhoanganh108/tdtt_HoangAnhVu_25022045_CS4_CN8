```python
#W7A1
def binary_search(a,n) :
    for i in range(len(a)) :
        if  a[i] == n :
            return i 
    return -1
if __name__ == '__main__':
    a = list(map(int,input().split()))
    n = int(input())
    print(binary_search(a,n))
```

```python
#W7A2
def count_occurrences(arr, x):
    cnt = 0
    for i in arr :
        if i == x:
            cnt += 1
    return cnt
if __name__ == '__main__' :
    a = list(map(int,input().split()))
    n = int(input())
    print(count_occurrences(a, n))
```

```python
#W7A3
def bubble_sort(arr) :
    n = len(arr)
    for i in range(n):
        for j in range(n-i-1):
            if arr[j] > arr[j + 1] :
                arr[j],arr[j + 1] = arr[j + 1],arr[j]
    return arr
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min = i
        for j in range(i + 1,n):
            if arr[j] < arr[min]:
                min = j
        arr[i],arr[min] = arr[min],arr[i]
    return arr
if __name__ == '__main__' :
    a = list(map(int,input().split()))
    print(bubble_sort(a))
    #print(selection_sort(a))
```

```python
#W7A13
a = list(map(int,input().split()))
kq = 0
for i in range(len(a)):
    tmp = a[i]
    dem = 1
    dem1 = 0
    dem2 = 0
    for j in range(i + 1,len(a)):
        if a[j] == a[i]:
            dem += 1
        if a[j] - a[i] == 1:
            dem1 += 1
        if a[j] - a[i] == -1:
            dem2 += 1
    m = max(dem + dem1,dem + dem2)
    kq = max(m,kq)
print(kq)
```
