```python
#4
s = input()
pairs = s.split()
d = {}
for p in pairs:
    key, value = p.split(":")
    if key in d:
      d[key] += [value]
    else :
      d[key] = [value]
print(d)
```

```python
#5
n = list(map(int,input().split()))
d = {'positives':0, 'negatives':0,'zeros':0 }
for i in n:
  if i == 0:
    d['zeros'] += 1
  elif i > 0:
    d['positives'] += 1
  else:
    d['negatives'] += 1
print(d)
```

```python
#9
import ast
a = input()
b = input()

d = ast.literal_eval(a)
D = ast.literal_eval(b)
for i in D:
  if i in d:
    d[i] += D[i]
  else:
    d[i] = D[i]
print(d)
```

```python
#10
n = list(map(int,input().split()))
m = int(input())
a = []
for i in range(len(n) - 1) :
  for j in range(i + 1,len(n)):
    if n[j] + n[i] == m and n[i] not in a:
      a.append(n[i])
a.sort()
b = []
for i in a:
  b.append((i,m - i))
print(b)
```

```python
#12
a = list(map(int,input().split()))
d = {}
for i in a :
  if i not in d :
    d[i] = 1
  else:
    d[i] += 1
m = 0
n = a[0]
for i in d :
  if d[i] > m :
     m = d[i]
     n = i
  elif d[i] == m:
    n = min(n,i)
print(n)
```
