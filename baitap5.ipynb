```python
#1
a,b = map(int,input().split())
def max_of_two(m,n):
  return max(m,n)
print(max_of_two(a,b))
```

```python
#3
def check(n):
  d = 0
  for i in range(1,n + 1):
    if n % i == 0 :
      d += 1
  if d == 2 :
    return True
  else:
    return False
n = int(input())
print(check(n))
```

```python
#4
def shh(x):
  s = 0
  for i in range (1,x):
    if x % i == 0:
      s += i
  if s == x:
    return True
  else:
    return False
n = int(input())
print(shh(n))
```

```python
#5
def check(a,k):
  m = -1
  for i in range(len(a)):
    if a[i] == k :
      m = i + 1
      break
  if m == -1:
    return -1
  else:
    return m
a = list(map(int,input().split()))
k = int(input())
print(check(a,k))

```

```python
#6
def giai_thua(n):
  s = 1
  for i in range (1,n+1):
    s *= i
  return s
n = int(input())
print(giai_thua(n))
```

```python
#7
def math(a,b,c):
  if b == "+" :
    return float(a) + float(c)
  elif b == "-":
    return float(a) - float(c)
  elif b == "*" :
    return float(a) * float(c)
  else :
    return float(a) / float(c)
a,b,c = map(str,input().split())
print('%.2f' % (math(a,b,c)))
```

```python
#8
def hamming_distance(x, y):
    ket_qua_xor = x ^ y
    so_luong_bit_khac = bin(ket_qua_xor).count('1')
    return so_luong_bit_khac
x,y = map(int,input().split())
print(hamming_distance(x,y))
```

```python
#9
def tong_cac_chu_so(n):
  s =  0
  while n > 0:
    s += n % 10
    n //= 10
  return s

n = int(input())
print(tong_cac_chu_so(n))
```

```python
#10
a,b = map(str,input().split())
def dang_cau(a,b):
    d = {}
    check = True
    for i in range(len(a)):
        if a[i] not in d :
            d[a[i]] = b[i]
        elif a[i] in d and d[a[i]] != b[i] :
            return False
            check = False
            break
    D = {}
    for i in range(len(b)):
        if b[i] not in D :
            D[b[i]] = a[i]
        elif b[i] in D and D[b[i]] != a[i] :
            return False
            check = False
            break
    if check == True :
        return True
print(dang_cau(a,b))
```
