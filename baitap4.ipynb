```python
#W4A8
n = int(input())
def check(x):
    s = 0
    while x > 0:
       s = s * 10 + x % 10
       x //= 10
    return s
dem = 0
while n != check(n):
    n += check(n)
    dem += 1
print(dem,n)
```

```python
#W4A32
s = input()
if len(s) <= 6:
    print("Mật khẩu yếu (độ dài phải > 6)")
else:
    co_hoa = False
    co_thuong = False
    co_so = False
    co_dac_biet = False

    for char in s:
        if char.isupper():
            co_hoa = True
        elif char.islower():
            co_thuong = True
        elif char.isdigit():
            co_so = True
        else:
            co_dac_biet = True

    if co_hoa and co_thuong and co_so and co_dac_biet:
        print("Đây là chuỗi mật khẩu mạnh")
    else:
        print("Đây là chuỗi mật khẩu không mạnh")
```
